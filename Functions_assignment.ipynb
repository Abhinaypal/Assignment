1. What is the difference between a function and a method in Python?

2. Explain the concept of function arguments and parameters in Python.

6. Explain the concept of generators in Python and how they are defined.
    - Generators in Python are a type of iterable that allow you to iterate through a sequence of values, but unlike lists or tuples, they do so one value at a time and only when requested. This makes them memory-efficient because they don’t store all values in memory at once.

      def count_up_to(max):
      count = 1
      while count <= max:
        yield count
        count += 1

      counter = count_up_to(5) # Using the generator

      for number in counter:
        print(number)


    - You define a generator using a function and the yield keyword. Each time yield is used, it produces a value and pauses the generator’s state, saving its local variables. When the generator is called again, it resumes right where it left off, continuing from the last yield statement. This makes generators perfect for handling large datasets or streams of data.


7. What are the advantages of using generators over regular functions?
    - Generators offer many advantages over regular functions:

    1. **Memory Efficiency**: Generators generate values on-the-fly without storing the entire data set in memory, which is especially useful for handling large datasets.

    2. **Lazy Evaluation**: They produce items only when requested, enabling efficient use of resources and faster execution time for operations that don't require all items at once.

    3. **Improved Performance**: Since values are produced one at a time, generators can perform better, particularly with large or infinite sequences.

    4. **Simplified Code**: Using yield simplifies code that handles iterative processing, making it more readable and easier to maintain.

      Generators help in writing efficient, readable, and resource-friendly code.

    - Purpose:
     Transform elements in an iterable by applying a specific function to each element.

    - Usage:
     You pass the function and the iterable(s) as arguments to map().

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum


user_input = input("Enter a list of numbers separated by spaces: ")
numbers = [int(num) for num in user_input.split()] #.split() for splitting the string by spaces

result = sum_of_even_numbers(numbers)
print("The sum of all even numbers is:", result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.





```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fib:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.current = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.n_terms:
            if self.current == 0:
                self.current += 1
                return self.a
            elif self.current == 1:
                self.current += 1
                return self.b
            else:
                self.current += 1
                self.a, self.b = self.b, self.a + self.b
                return self.b
        else:
            raise StopIteration

n_terms = int(input("Enter the number of terms: "))
fib_iterator = Fib(n_terms)

for num in fib_iterator:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    num = 0
    while num <= exponent:
        yield 2 ** num
        num += 1

exponent = int(input("Enter the maximum exponent: "))
for i in powers_of_two(exponent):
    print(i)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

file_path = 'file.txt'  # Replace with your file path
for line in read_file(file_path):
    print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
tuples_list = [(1, 3), (4, 1), (5, 2), (7, 0)]

sorted_list = sorted(tuples_list, key=lambda x: x[1]) #sort the list by sorted() method using lambda function

print(sorted_list)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def cel_to_fahr(celsius):
    return (celsius * 9/5) + 32  #using the formula of celsius to fahrenheit
  #F = (c*9/5)+32
celsius_list = [0, 20, 37, 100]

fahrenheit = list(map(cel_to_fahr, celsius_list))

print(fahrenheit)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    return char.lower() not in "aeiou"
input_string = input("Enter your sentence/word: ")

result = "".join(filter(is_not_vowel, input_string))  #using filter() to remove vowel

print("Your sentence/word after removing vowels:", result)
```
