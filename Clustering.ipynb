# Theoretical Questions

1.  What is unsupervised learning in the context of machine learning
  - Unsupervised learning is a type of machine learning where the algorithm learns patterns from **data that has no labels**. It tries to group, organize, or compress the data based on similarities or structures—like clustering customers by behavior or reducing data dimensions to highlight key features.


2.  How does K-Means clustering algorithm work
    - K-Means is a clustering algorithm that groups similar data points into **K number of clusters**.

  Here’s how it works, in a nutshell:
  1. Choose the number of clusters, *K*.
  2. Randomly place *K* centroids (cluster centers).
  3. Assign each data point to the nearest centroid.
  4. Recalculate centroids based on the assigned points.
  5. Repeat steps 3–4 until the centroids stop changing much.

  It’s simple, fast, and widely used for grouping and pattern discovery.



3.  Explain the concept of a dendrogram in hierarchical clustering
  - A **dendrogram** is a tree-like diagram used in hierarchical clustering to show how data points are grouped together step by step.

  Each branch represents a cluster merging event, starting with individual points at the bottom and merging upward until all points form one cluster at the top. The height of the branches reflects the **distance or similarity** between clusters when they were joined.




4.  What is the main difference between K-Means and Hierarchical Clustering
  - The main difference is in their **approach and structure**:

  - **K-Means** requires you to pre-define the number of clusters (*K*) and uses a flat, non-hierarchical structure.
  - **Hierarchical Clustering** builds a tree (dendrogram) of clusters without needing to set the number upfront, and it shows how clusters are formed step-by-step.

  In short: *K-Means is fast and flat; Hierarchical is slower but builds a nested structure.*



5.  What are the advantages of DBSCAN over K-Means
  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has a few key advantages over K-Means:

  - **No need to predefine cluster count**—DBSCAN finds clusters based on data density.
  - **Can find clusters of arbitrary shape**, not just circular ones like K-Means.
  - **Handles noise and outliers well** by labeling low-density points as noise.
  - **More robust** when clusters have varied sizes and densities.

  It’s especially useful when you don't know how many clusters to expect or when your data isn’t neatly separable.




6.  When would you use Silhouette Score in clustering
  - You’d use the **Silhouette Score** to **evaluate how well your clustering algorithm performed**.

  It measures how similar each point is to its own cluster compared to other clusters. A higher score (closer to 1) means **better-defined, well-separated clusters**. It's great for choosing the optimal number of clusters or comparing different clustering methods.



7.  What are the limitations of Hierarchical Clustering
  - Here are a few key limitations of Hierarchical Clustering:

  - **Scalability**: It’s computationally expensive and doesn’t scale well with large datasets.
  - **No backtracking**: Once a merge or split is made, it can't be undone—even if it leads to poor clustering.
  - **Choice of distance metric and linkage** can significantly affect the outcome.
  - **Sensitive to noise and outliers**, which can distort the cluster structure.

  Despite these, it’s still powerful for understanding data structure when working with smaller, well-behaved datasets.



8.  Why is feature scaling important in clustering algorithms like K-Means
  - Feature scaling is important in K-Means because it **ensures all features contribute equally** to the distance calculations used to form clusters.

  Without scaling, features with larger ranges (like income in thousands vs. age in years) can dominate, leading to biased and misleading clusters. Scaling levels the playing field, giving each feature a fair say.



9.  How does DBSCAN identify noise points
  - DBSCAN identifies **noise points** as those that **don’t belong to any cluster**.

  Specifically, if a point has **fewer neighbors than a set threshold (MinPts)** within a given radius (ε), and it's **not close enough to any core point**, it’s labeled as noise or an outlier. These points are left unclustered.



10.  Define inertia in the context of K-Means
  - In K-Means, **inertia** is a measure of how tightly the data points are clustered around their centroids.

  It’s calculated as the **sum of squared distances** between each point and its assigned cluster center. Lower inertia means more compact, well-formed clusters—but too low might also mean overfitting. It’s commonly used to help choose the best number of clusters.

11.  What is the elbow method in K-Means clustering
  - The **elbow method** is a technique used to find the **optimal number of clusters (K)** in K-Means.

  It works by plotting the *inertia* (within-cluster sum of squares) for different K values. As K increases, inertia decreases—but after a certain point, the rate of decrease slows down, forming an “elbow” in the plot. That elbow point suggests a good balance between model complexity and fit.


12.  Describe the concept of "density" in DBSCAN
  - In DBSCAN, **density** refers to how many data points exist within a specified radius (*ε*) around a point.

  A region is considered **dense** if it has at least a minimum number of points (*MinPts*) within that radius. DBSCAN uses this concept to identify **core points**, build clusters, and distinguish noise.


13.  Can hierarchical clustering be used on categorical data
  -  Yes, **hierarchical clustering can be used on categorical data**, but with some adjustments.

  Since it relies on a distance measure, you’d need to use **appropriate metrics** like *Hamming distance* or *Jaccard similarity* for categorical variables. It works best when paired with methods that handle non-numeric data effectively.


14.  What does a negative Silhouette Score indicate
  - A **negative Silhouette Score** means a data point is likely assigned to the wrong cluster—it’s **closer to a neighboring cluster** than to its own.

  In short, it signals **poor clustering**, suggesting the structure doesn’t fit the data well. Time to reconsider the number of clusters or clustering method!



15.  Explain the term "linkage criteria" in hierarchical clustering
  - **Linkage criteria** determine how the distance between clusters is calculated in hierarchical clustering.

  Common types include:
  - **Single linkage**: shortest distance between any two points in the clusters.
  - **Complete linkage**: farthest distance between any two points.
  - **Average linkage**: average distance between all point pairs.
  - **Ward’s method**: minimizes variance within clusters.

  The choice of linkage affects the shape and structure of the resulting dendrogram.



16.  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
  - K-Means can perform poorly on data with **varying cluster sizes or densities** because it assumes:

  - **All clusters are similar in size and density**.
  - **Clusters are spherical and equally spaced**.

  So, when actual clusters are uneven, K-Means may split large clusters, merge small ones, or misplace boundaries—leading to inaccurate results. It's not well-suited for complex shapes or varied distributions.



17.  What are the core parameters in DBSCAN, and how do they influence clustering
  - The two core parameters in **DBSCAN** are:

  - **ε (epsilon)**: The radius around a point to search for neighbors. A larger ε means broader neighborhoods and potentially larger clusters.
  - **MinPts**: The minimum number of points required within ε to form a dense region (i.e., a core point).

  These control how **clusters are defined based on density**—tune them carefully to avoid merging distinct clusters or labeling true points as noise.


18.  How does K-Means++ improve upon standard K-Means initialization
  - K-Means++ improves K-Means by **smarter initialization of centroids**.

  Instead of random selection, it picks the first centroid randomly, then chooses subsequent ones with a preference for data points **far from existing centroids**. This spreads the initial centroids out, reducing chances of poor clustering and helping K-Means converge faster and more accurately.


19.  What is agglomerative clustering
  - Agglomerative clustering is a **bottom-up hierarchical clustering** method.
  It starts with each data point as its own cluster, then **repeatedly merges the closest pairs of clusters** based on a linkage criterion (like single, complete, or average linkage) until all points are merged into a single cluster or a desired number of clusters is reached. It’s intuitive and doesn’t need you to predefine cluster count.



20.  What makes Silhouette Score a better metric than just inertia for model evaluation?
  - The **Silhouette Score** is considered better than just using inertia for clustering evaluation because it gives a more **balanced and interpretable assessment** of cluster quality. Here's why:

  - **Inertia only decreases** as you add more clusters—it doesn’t tell you if the clusters are well-separated or meaningful.
  - **Silhouette Score considers both cohesion and separation**: how close each point is to its own cluster vs. other clusters.
  - It provides a score between **-1 and 1**, making it easier to compare across different models or cluster counts.

```python
# 1. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot




import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the results
plt.figure(figsize=(8, 6))

# Plot data points with cluster colors
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data points')

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')

# Add labels and title
plt.title('K-Means Clustering on Synthetic Data (4 Centers)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 9.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart



# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with 4 centers
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Initialize lists to store silhouette scores
k_values = range(2, 6)
silhouette_scores = []

# Apply K-Means and compute silhouette scores for k = 2 to 5
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue', edgecolor='black')
plt.title('Silhouette Scores for K-Means Clustering (k = 2 to 5)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print silhouette scores
print("Silhouette Scores:")
for k, score in zip(k_values, silhouette_scores):
    print(f"k = {k}: {score:.4f}")
```

```python
# 11. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with overlapping clusters
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Create a mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

# Predict cluster labels for each point in the mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Visualize the clusters and decision boundaries
plt.figure(figsize=(8, 6))

# Plot decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k', alpha=0.8)

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')

# Add labels and title
plt.title('K-Means Clustering with Decision Boundaries (Overlapping Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 14. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot


# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Initialize list to store inertia values
k_values = range(2, 7)
inertia_values = []

# Apply K-Means and compute inertia for k = 2 to 6
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='b')
plt.title('Inertia Values for K-Means Clustering (k = 2 to 6) on Breast Cancer Dataset')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.xticks(k_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Print inertia values
print("Inertia Values:")
for k, inertia in zip(k_values, inertia_values):
    print(f"k = {k}: {inertia:.2f}")

```

```python
# 23. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Set random seed for reproducibility
np.random.seed(42)

# Generate noisy circular data
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.1, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Create side-by-side scatter plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# K-Means plot
ax1.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=50, cmap='viridis', edgecolor='k', alpha=0.8)
ax1.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
ax1.set_title('K-Means Clustering')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# DBSCAN plot
scatter = ax2.scatter(X[:, 0], X[:, 1], c=dbscan_labels, s=50, cmap='viridis', edgecolor='k', alpha=0.8)
ax2.set_title('DBSCAN Clustering')
ax2.set_xlabel('Feature 1')
ax2.grid(True, linestyle='--', alpha=0.7)
if -1 in dbscan_labels:
    ax2.scatter(X[dbscan_labels == -1, 0], X[dbscan_labels == -1, 1], c='red', s=50, marker='x', label='Outliers')
ax2.legend()

plt.tight_layout()
plt.show()

# Print number of DBSCAN outliers
n_outliers = np.sum(dbscan_labels == -1)
print(f"Number of outliers detected by DBSCAN: {n_outliers}")
```

```python
# 24.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform KMeans clustering
n_clusters = 3  # Iris has 3 classes
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Compute silhouette scores
silhouette_avg = silhouette_score(X, cluster_labels)
sample_silhouette_values = silhouette_samples(X, cluster_labels)

# Create a silhouette plot
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
y_lower = 10

# Plot silhouette coefficients for each cluster
for i in range(n_clusters):
    # Aggregate the silhouette scores for samples in cluster i
    cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    cluster_silhouette_values.sort()
    size_cluster_i = cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # Fill the silhouette plot
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, cluster_silhouette_values,
                     facecolor=f'C{i}', edgecolor=f'C{i}', alpha=0.7)

    # Label the silhouette plots with cluster numbers
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # Gap between clusters

# Add details to the plot
ax.set_title("Silhouette Plot for Iris Dataset (KMeans Clustering)")
ax.set_xlabel("Silhouette Coefficient")
ax.set_ylabel("Cluster Label")
ax.axvline(x=silhouette_avg, color="red", linestyle="--", label=f"Average Silhouette Score: {silhouette_avg:.3f}")
ax.legend()
ax.set_yticks([])  # Remove y-axis ticks
ax.set_xlim([-0.1, 1])
ax.set_ylim([0, len(X) + (n_clusters + 1) * 10])

plt.show()
```

```python
# 27. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count



import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
n_samples = 300
n_clusters = 3
X, _ = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=1.0, random_state=42)

# Add noise by generating additional random points
n_noise = 50
noise = np.random.uniform(low=X.min(axis=0), high=X.max(axis=0), size=(n_noise, 2))
X_noisy = np.vstack([X, noise])

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
cluster_labels = dbscan.fit_predict(X_noisy)

# Count clusters and noise points
n_clusters_found = len(set(cluster_labels) - {-1})  # Exclude noise label (-1)
n_noise_points = np.sum(cluster_labels == -1)

# Print results
print(f"Number of clusters found: {n_clusters_found}")
print(f"Number of noise points: {n_noise_points}")
```
