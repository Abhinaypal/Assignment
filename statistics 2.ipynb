# Theory questions

 1.  What is hypothesis testing in statistics?
  - Hypothesis testing in statistics is a method used to determine if there is enough evidence in a sample of data to support a certain claim about a population. It involves making an assumption (hypothesis), collecting data, and using statistical techniques to analyze whether the assumption is likely true or should be rejected. It helps in making data-driven decisions.


2.  What is the null hypothesis, and how does it differ from the alternative hypothesis
  - The **null hypothesis (H₀)** is a statement that assumes no effect or no difference in a study—it suggests that any observed result is due to chance.  

  - The **alternative hypothesis (H₁)** is the opposite, stating that there is an actual effect or difference.  

  - Hypothesis testing aims to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.



3.  What is the significance level in hypothesis testing, and why is it important
  - The **significance level (α)** in hypothesis testing is the probability of rejecting the **null hypothesis** when it is actually true. Common values are **0.05 or 0.01**, meaning there's a **5% or 1% risk** of making an incorrect conclusion.  

  - It is important because it sets the threshold for deciding whether an observed result is statistically significant, helping researchers avoid false conclusions!


4. What does a P-value represent in hypothesis testing
  - A **P-value** in hypothesis testing represents the probability of obtaining the observed results, or more extreme ones, if the **null hypothesis (H₀)** is true.  

  - A **small P-value (typically ≤ 0.05)** suggests strong evidence against the null hypothesis, leading to its rejection. A **large P-value** means there is not enough evidence to reject it.  



5. How do you interpret the P-value in hypothesis testing
  - Interpreting the **P-value** in hypothesis testing depends on its value:  

  - **Small P-value (≤ 0.05):** Strong evidence against the **null hypothesis (H₀)**, so you **reject H₀**—suggesting the result is **statistically significant**.  
  - **Large P-value (> 0.05):** Weak evidence against **H₀**, meaning you **fail to reject H₀**—suggesting the result **might be due to chance**.  

  -Essentially, the **smaller the P-value, the stronger the evidence** against the null hypothesis!



6.  What are Type 1 and Type 2 errors in hypothesis testing
  - In hypothesis testing:

  - **Type 1 error (False Positive):** Rejecting a true null hypothesis. It means concluding there is an effect when none exists.
  - **Type 2 error (False Negative):** Failing to reject a false null hypothesis. It means missing an actual effect when one exists.



7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
  - One-tailed test: Tests for an effect in only one direction (either greater or smaller than a certain value).

  - Two-tailed test: Tests for an effect in both directions (greater or smaller), checking for any significant difference.


8. What is the Z-test, and when is it used in hypothesis testing
  - The **Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means when the population variance is known.

  It is used when:
  - The sample size is large (typically **n ≥ 30**).
  - The data follows a **normal distribution**.
  - The population **variance is known**.

  It helps assess hypotheses about population parameters using the standard normal distribution!



9. How do you calculate the Z-score, and what does it represent in hypothesis testing.
  - The z score is calculated as:
    Z = (X-μ)/σ

  - where X = sample value
    mu = population mean
    sigma = population standard deviation

  - It represents how many standard deciations a data point is from the mean, helping statistical significance in hypothesis testing.


10.  What is the T-distribution, and when should it be used instead of the normal distribution
  - The **T-distribution** is a probability distribution used when estimating population parameters when the sample size is **small** or the **population standard deviation is unknown**.

  It is used instead of the normal distribution when:
  - **Sample size** is **small (n < 30)**.  
  - **Population variance** is **unknown**.  
  - Data is **approximately normal** but has more variability due to a smaller sample.

  It helps make more reliable inferences when dealing with limited data!


11.  What is the difference between a Z-test and a T-test?
  - The **Z-test** and **T-test** are both used for hypothesis testing, but they differ in key ways:

  - **Z-test:** Used when the **sample size is large (n ≥ 30)** and the **population variance is known**. It relies on the **normal distribution**.  
  - **T-test:** Used when the **sample size is small (n < 30)** or the **population variance is unknown**. It relies on the **T-distribution**, which accounts for greater variability in small samples.

  Choosing between them depends on sample size and whether population variance is known!



12.  What is the T-test, and how is it used in hypothesis testing
  - The **T-test** is a statistical test used to compare **means** between groups when the **sample size is small** or the **population variance is unknown**.

  It is used in hypothesis testing to determine if there is a **significant difference** between:
  - A sample mean and a population mean (**One-sample T-test**).
  - Two independent groups (**Independent T-test**).
  - Two related samples (**Paired T-test**).

  The T-test relies on the **T-distribution** and helps assess whether observed differences are **due to chance or actual effects**.


13.  What is the relationship between Z-test and T-test in hypothesis testing
  - The **Z-test** and **T-test** are both used to test hypotheses about population means, but they differ in their assumptions:

  - **Z-test:** Used when the **sample size is large (n ≥ 30)** and the **population variance is known**. It follows the **normal distribution**.  
  - **T-test:** Used when the **sample size is small (n < 30)** or the **population variance is unknown**. It follows the **T-distribution**, which has heavier tails to account for more variability.

  Both tests assess whether observed differences are statistically significant, but the choice depends on sample size and variance availability!



14.  What is a confidence interval, and how is it used to interpret statistical results
  - A **confidence interval (CI)** is a range of values that estimates a population parameter with a certain level of confidence (e.g., **95% or 99%**).

  It helps interpret statistical results by:
  - Indicating the **range** within which the true population value likely falls.
  - Showing **uncertainty** in an estimate—wider intervals mean more variability.
  - Helping determine **significance**—if a CI excludes a certain value (e.g., zero in a mean difference), the result is likely **statistically significant**.

  Confidence intervals provide a more informative alternative to simple hypothesis tests!



15.  What is the margin of error, and how does it affect the confidence interval
  - The **margin of error (MOE)** is the maximum expected difference between the sample estimate and the true population value.

  It affects the **confidence interval (CI)** by:
  - **Determining width**—larger MOE results in a wider CI, indicating more uncertainty.
  - **Depending on sample size**—larger samples reduce MOE, leading to a more precise estimate.
  - **Depending on confidence level**—higher confidence (e.g., **99% vs. 95%**) increases MOE, making the CI wider.

  A smaller MOE provides a more precise statistical estimate!



16. How is Bayes' Theorem used in statistics, and what is its significance
  - **Bayes' Theorem** is used in statistics to update the probability of an event occurring based on new evidence.

  It is significant because:
  - It helps in **conditional probability**—determining the likelihood of an event given prior knowledge.
  - It is widely used in **machine learning, medical diagnosis, and spam filtering**.
  - It improves **decision-making** by incorporating new data into probability estimates.

  Bayes' Theorem allows for more dynamic and adaptive statistical reasoning!



17.  What is the Chi-square distribution, and when is it used
  - The **Chi-square distribution** is a probability distribution used in statistical tests involving categorical data.

  It is used when:
  - **Testing independence** between two categorical variables (**Chi-square test for independence**).  
  - **Goodness-of-fit** analysis to check how well observed data matches expected values.  
  - **Variance analysis** in population studies.

  It is widely applied in hypothesis testing for categorical data relationships!


18.  What is the Chi-square goodness of fit test, and how is it applied
  - The **Chi-square goodness-of-fit test** is used to determine if observed categorical data fits an expected distribution.

  It is applied by:
  - Comparing observed frequencies with expected frequencies.
  - Calculating the **Chi-square statistic** to measure the deviation.
  - Checking significance using the **Chi-square distribution** and a **p-value**.

  It helps assess whether data aligns with a theoretical model or expected proportions!


19.  What is the F-distribution, and when is it used in hypothesis testing
  - The **F-distribution** is a probability distribution used in statistical analysis, mainly in **variance comparison**.

  It is used in hypothesis testing when:
  - **Comparing variances** between two datasets (**F-test**).  
  - **ANOVA (Analysis of Variance)** to test if multiple group means are significantly different.  
  - **Regression analysis** to assess model fit.

  Since the F-distribution is **skewed and non-negative**, it is particularly useful for variance-based tests!


20.  What is an ANOVA test, and what are its assumptions
  - **ANOVA (Analysis of Variance)** is a statistical test used to compare the means of multiple groups to determine if they are significantly different.

  Its assumptions include:
  - **Normality:** Data should follow a normal distribution.  
  - **Independence:** Observations must be independent of each other.  
  - **Homogeneity of variance:** The variances of different groups should be approximately equal.

  ANOVA helps analyze differences across multiple groups efficiently!


21. What are the different types of ANOVA tests
  - There are several types of **ANOVA (Analysis of Variance) tests**, each designed for specific scenarios:

  1. **One-Way ANOVA:** Compares the means of **three or more** independent groups based on **one** factor.  
  2. **Two-Way ANOVA:** Analyzes the effect of **two** independent factors on a dependent variable.  
  3. **Repeated Measures ANOVA:** Used when the same subjects are measured multiple times under different conditions.  
  4. **MANOVA (Multivariate ANOVA):** Extends ANOVA to multiple dependent variables instead of just one.  

  These tests help determine whether group means differ significantly in different experimental settings!


22.  What is the F-test, and how does it relate to hypothesis testing?
  - The **F-test** is a statistical test used to compare **variances** between groups.

  It relates to hypothesis testing by:
  - **Assessing equality of variances** in two or more populations.  
  - Being used in **ANOVA (Analysis of Variance)** to check for significant differences among group means.  
  - Evaluating **regression models** to determine if explanatory variables significantly impact the dependent variable.

  Since the **F-distribution** is skewed and non-negative, the F-test is useful for testing variance-based hypotheses!

```python
# 15.  Implement hypothesis testing using Z-statistics for a sample dataset


import numpy as np
import scipy.stats as stats
def z_test(sample_data, population_mean, population_std_dev):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Example usage
sample_data = np.random.normal(50, 10, 30)  # Sample dataset
population_mean = 50  # Hypothesized mean
population_std_dev = 10  # Known population standard deviation

z_score, p_value = z_test(sample_data, population_mean, population_std_dev)
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis (Significant difference).")
else:
    print("Fail to reject the null hypothesis (No significant difference).")
```

```python
# 16.  Create a confidence interval for a dataset using Python and interpret the result



def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    sample_size = len(data)

    # Compute margin of error
    margin_of_error = stats.t.ppf((1 + confidence) / 2, sample_size - 1) * (sample_std_dev / np.sqrt(sample_size))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
data = np.random.normal(50, 10, 30)  # Sample dataset
ci_lower, ci_upper = confidence_interval(data)
print(f"Confidence Interval (95%): ({ci_lower:.2f}, {ci_upper:.2f})")

# Interpretation
print("This means that we are 95% confident that the true population mean falls within this interval.")
```

```python
# 21.  Write a Python program to compare the standard deviations of two datasets and visualize the difference


import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviation(data1, data2):
    std_dev1 = np.std(data1, ddof=1)
    std_dev2 = np.std(data2, ddof=1)

    # Plot histogram of both datasets
    plt.hist(data1, bins=30, alpha=0.5, label=f"Dataset 1 (std = {std_dev1:.2f})", color='blue', edgecolor='black')
    plt.hist(data2, bins=30, alpha=0.5, label=f"Dataset 2 (std = {std_dev2:.2f})", color='red', edgecolor='black')
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Comparison of Standard Deviations")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return std_dev1, std_dev2

# Example usage
data1 = np.random.normal(50, 10, 1000)  # Dataset with std dev 10
data2 = np.random.normal(50, 20, 1000)  # Dataset with std dev 20

std_dev1, std_dev2 = compare_standard_deviation(data1, data2)
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results


import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std_dev):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Compute Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Example usage
sample_data = np.random.normal(50, 10, 30)  # Generating a random sample
population_mean = 50  # Hypothesized population mean
population_std_dev = 10  # Known population standard deviation

z_score, p_value = z_test(sample_data, population_mean, population_std_dev)
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
# 2.  Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python


import numpy as np
import scipy.stats as stats

def hypothesis_test(sample_data, population_mean, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    # Compute t-statistic
    t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=sample_size - 1))

    return t_statistic, p_value

# Simulate random data
np.random.seed(42)  # Setting seed for reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Generating normally distributed data

# Hypothesized population mean
population_mean = 50

# Perform hypothesis test
t_statistic, p_value = hypothesis_test(sample_data, population_mean)

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
# 5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)

    # Null hypothesis distribution (mean = 0, std = 1)
    null_pdf = stats.norm.pdf(x, 0, 1)

    # Alternative hypothesis distribution (mean = 2, std = 1)
    alt_pdf = stats.norm.pdf(x, 2, 1)

    # Critical value for significance level
    critical_value = stats.norm.ppf(1 - alpha)

    plt.plot(x, null_pdf, 'b-', lw=2, label="Null Hypothesis (H0)")
    plt.plot(x, alt_pdf, 'r-', lw=2, label="Alternative Hypothesis (H1)")

    # Shade Type 1 error (rejecting H0 when it's true)
    plt.fill_between(x, null_pdf, where=(x > critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 error (failing to reject H0 when H1 is true)
    plt.fill_between(x, alt_pdf, where=(x < critical_value), color='purple', alpha=0.3, label="Type 2 Error (β)")

    plt.axvline(critical_value, color='black', linestyle='--', label="Critical Value")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
visualize_type1_type2_errors()
```

```python
# 6.  Write a Python program to perform an independent T-test and interpret the results


import numpy as np
import scipy.stats as stats

def independent_t_test(data1, data2, alpha=0.05):
    t_statistic, p_value = stats.ttest_ind(data1, data2, equal_var=False)  # Welch’s t-test (unequal variance)

    return t_statistic, p_value

# Generate sample data
np.random.seed(42)  # Setting seed for reproducibility
data1 = np.random.normal(loc=50, scale=10, size=30)  # First dataset (Mean=50, Std=10)
data2 = np.random.normal(loc=55, scale=10, size=30)  # Second dataset (Mean=55, Std=10)

# Perform independent t-test
t_statistic, p_value = independent_t_test(data1, data2)

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The means of the two datasets are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the means of the two datasets.")
```

```python
# 9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance



import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    sample_size = len(data)

    # Compute margin of error
    margin_of_error = stats.t.ppf((1 + confidence) / 2, sample_size - 1) * (sample_std_dev / np.sqrt(sample_size))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
data = np.random.normal(50, 10, 30)  # Sample dataset
ci_lower, ci_upper = confidence_interval(data)
print(f"Confidence Interval (95%): ({ci_lower:.2f}, {ci_upper:.2f})")

# Interpretation
print("This means that we are 95% confident that the true population mean falls within this interval.")
```

```python
# 16.  Implement an F-test using Python to compare the variances of two random samples


import numpy as np
import scipy.stats as stats

# Generate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=10, scale=5, size=30)  # Mean=10, Std=5
sample2 = np.random.normal(loc=10, scale=8, size=30)  # Mean=10, Std=8

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform F-test
F_statistic = var1 / var2
df1, df2 = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom
p_value = stats.f.cdf(F_statistic, df1, df2) * 2  # Two-tailed test

# Print results
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
# 19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    print("\n### Normality Test (Shapiro-Wilk) ###")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: p-value = {p_value:.4f} {'(Reject H0: Not normal)' if p_value < 0.05 else '(Fail to reject H0: Normal)'}")

    # Independence Check (Visual Inspection using Box Plot)
    print("\n### Independence Check (Visual Inspection) ###")
    plt.figure(figsize=(8, 5))
    plt.boxplot(groups, labels=[f"Group {i+1}" for i in range(len(groups))])
    plt.title("Independence Check: Box Plot of Groups")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

    # Equal Variance Check (Levene's Test)
    print("\n### Equal Variance Test (Levene’s Test) ###")
    stat, p_value = stats.levene(*groups)
    print(f"Levene’s Test p-value = {p_value:.4f} {'(Reject H0: Variances not equal)' if p_value < 0.05 else '(Fail to reject H0: Variances are equal)'}")

# Example usage with sample data
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results


import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
np.random.seed(42)
data = {
    "Factor1": np.repeat(["A", "B"], 30),
    "Factor2": np.tile(["X", "Y"], 30),
    "Response": np.random.normal(loc=50, scale=10, size=60)
}

# Create DataFrame
df = pd.DataFrame(data)

# Perform Two-Way ANOVA
model = smf.ols("Response ~ C(Factor1) * C(Factor2)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA results
print(anova_table)

# Visualization of interaction effect
plt.figure(figsize=(8, 5))
sns.boxplot(x="Factor1", y="Response", hue="Factor2", data=df)
plt.title("Interaction between Two Factors")
plt.xlabel("Factor1")
plt.ylabel("Response")
plt.legend(title="Factor2")
plt.grid()
plt.show()
```

```python
# 22.  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate sample data for three groups
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Perform one-way ANOVA test
F_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print results
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Plot the distributions of the groups
plt.figure(figsize=(8, 5))
plt.boxplot([group1, group2, group3], labels=["Group 1", "Group 2", "Group 3"])
plt.title("Comparison of Group Means (ANOVA)")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
# 23.  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means



import numpy as np
import scipy.stats as stats

# Generate random data from a normal distribution
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between means.")

```

```python
# 24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results



import numpy as np
import scipy.stats as stats

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5

# Define population variance (hypothetical)
population_variance = 25  # Given variance (σ^2)

# Compute sample variance
sample_variance = np.var(sample, ddof=1)
n = len(sample)

# Compute Chi-square test statistic
chi2_stat = (n - 1) * sample_variance / population_variance

# Compute p-value
p_value = stats.chi2.sf(chi2_stat, df=n-1)  # Upper tail test

# Print results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variance significantly differs from the population variance.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variance.")
```

```python
# 25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_proportions(successes1, total1, successes2, total2):
    # Compute proportions
    p1 = successes1 / total1
    p2 = successes2 / total2

    # Compute pooled proportion
    p_pool = (successes1 + successes2) / (total1 + total2)

    # Compute standard error
    se = np.sqrt(p_pool * (1 - p_pool) * (1 / total1 + 1 / total2))

    # Compute Z-statistic
    z_stat = (p1 - p2) / se

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    return z_stat, p_value

# Example usage
successes_group1 = 50
total_group1 = 200
successes_group2 = 30
total_group2 = 180

z_stat, p_value = z_test_proportions(successes_group1, total_group1, successes_group2, total_group2)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between proportions.")
```

```python
# 26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=10, scale=5, size=30)  # Mean=10, Std=5
sample2 = np.random.normal(loc=10, scale=8, size=30)  # Mean=10, Std=8

# Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform F-test
F_statistic = var1 / var2
df1, df2 = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom
p_value = 2 * (1 - stats.f.cdf(F_statistic, df1, df2))  # Two-tailed test

# Print results
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8, 5))
plt.boxplot([sample1, sample2], labels=["Sample 1", "Sample 2"])
plt.title("Comparison of Variances (F-test)")
plt.xlabel("Samples")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
# 27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.



import numpy as np
import scipy.stats as stats

# Simulated observed data (frequencies)
observed = np.array([45, 35, 20])

# Expected distribution (assuming equal proportions)
total_count = np.sum(observed)
expected = np.array([total_count / 3] * 3)

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution significantly differs from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")
```
