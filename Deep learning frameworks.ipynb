16.  What are some common activation functions used in neural networks
  - üß† **Common Activation Functions in Neural Networks**:

  - **ReLU (Rectified Linear Unit)**: `f(x) = max(0, x)` ‚Äì Fast and widely used in hidden layers.
  - **Sigmoid**: `f(x) = 1 / (1 + e^-x)` ‚Äì Good for binary classification.
  - **Tanh**: `f(x) = (e^x - e^-x) / (e^x + e^-x)` ‚Äì Outputs between -1 and 1.
  - **Softmax**: Converts outputs into probabilities for multi-class classification.
  - **Leaky ReLU**: Allows small gradient for negative inputs to fix ‚Äúdying ReLU‚Äù.
  - **Swish**: `f(x) = x * sigmoid(x)` ‚Äì Smooth and effective in deep networks.
  - **ELU (Exponential Linear Unit)**: Like ReLU but smoother for negative values.



17.  What is the difference between torch.nn.Module and torch.nn.Sequential in PyTorch
  - **Quick Difference: `nn.Module` vs `nn.Sequential`**

  - nn.Module: Base class for all models. You define layers and a custom forward() method.
  - nn.Sequential: A shortcut for stacking layers in order. No need to write forward() manually.

  Use `nn.Module` for **flexible architectures**.  
  Use `nn.Sequential` for **simple, linear models**.

```python
# 4.  How can you visualize the training progress using TensorFlow and Matplotlib



import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Define the neural network (same as previous example)
def create_simple_nn(input_shape, hidden_units, output_units):
    model = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Dense(hidden_units, activation='relu'),
        layers.Dense(output_units, activation='sigmoid')  # Binary classification
    ])
    return model

# Create and compile the model
input_shape = (10,)  # 10 input features
hidden_units = 64
output_units = 1
model = create_simple_nn(input_shape, hidden_units, output_units)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Generate dummy data
X_train = np.random.random((100, 10))  # 100 samples, 10 features
y_train = np.random.randint(0, 2, (100, 1))  # Binary labels
X_val = np.random.random((20, 10))  # Validation data
y_val = np.random.randint(0, 2, (20, 1))  # Validation labels

# Train the model and store training history
history = model.fit(X_train, y_train, epochs=10, batch_size=32,
                    validation_data=(X_val, y_val), verbose=1)

# Plot training and validation loss/accuracy
def plot_training_history(history):
    # Extract metrics from history
    loss = history.history['loss']
    val_loss = history.history.get('val_loss', [])
    accuracy = history.history['accuracy']
    val_accuracy = history.history.get('val_accuracy', [])

    epochs = range(1, len(loss) + 1)

    # Plot loss
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, 'b-', label='Training Loss')
    if val_loss:
        plt.plot(epochs, val_loss, 'r-', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, accuracy, 'b-', label='Training Accuracy')
    if val_accuracy:
        plt.plot(epochs, val_accuracy, 'r-', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_training_history(history)
```

```python
# 6.  How do you create a simple neural network in PyTorch



import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(10, 20)  # Input: 10, Hidden: 20
        self.relu = nn.ReLU()         # Activation function
        self.fc2 = nn.Linear(20, 2)   # Hidden: 20, Output: 2

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Check for GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Create model instance
model = SimpleNN().to(device)
```

```python
criterion = nn.CrossEntropyLoss()  # For classification
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# Generate dummy data
torch.manual_seed(42)  # For reproducibility
X = torch.randn(100, 10).to(device)  # 100 samples, 10 features
y = torch.randint(0, 2, (100,)).to(device)  # Binary labels (0 or 1)
```

```python
# Training loop
num_epochs = 100
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X)
    loss = criterion(outputs, y)

    # Backward pass and optimization
    optimizer.zero_grad()  # Clear previous gradients
    loss.backward()        # Compute gradients
    optimizer.step()       # Update weights

    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# Test with a single sample
model.eval()  # Set to evaluation mode
with torch.no_grad():
    test_input = torch.randn(1, 10).to(device)  # Single sample
    prediction = model(test_input)
    predicted_class = torch.argmax(prediction, dim=1)
    print(f"Predicted class: {predicted_class.item()}")
```

```python
# 8.  How do you implement a custom loss function in PyTorch


import torch

def mape_loss(output, target):
    """
    Mean Absolute Percentage Error Loss
    output: Predicted values (tensor)
    target: True values (tensor)
    """
    epsilon = 1e-6  # Avoid division by zero
    absolute_error = torch.abs(output - target)
    percentage_error = absolute_error / (torch.abs(target) + epsilon)
    return torch.mean(percentage_error) * 100  # Return MAPE as percentage
```

```python
# Dummy data
predictions = torch.tensor([2.5, 0.0, 2.1, 7.8])
targets = torch.tensor([3.0, 0.5, 2.0, 7.5])

# Compute loss
loss = mape_loss(predictions, targets)
print(f"MAPE Loss: {loss.item():.4f}%")
```
