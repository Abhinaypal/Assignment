# Theoritical questions

1.  What is Boosting in Machine Learning
  - Boosting is a machine learning technique used to improve the accuracy of models by combining the outputs of several weaker models (often decision trees) into one strong model. Each new model focuses on correcting the errors made by the previous ones, so the final result is better at making predictions. It's like a team where each member learns from the mistakes of the others to get stronger together.





2.  How does Boosting differ from Bagging
  - Boosting and Bagging are both ensemble learning techniques, but they differ in how models are built:

  - **Bagging (Bootstrap Aggregating)** builds multiple models independently and in parallel, using random subsets of the data. It reduces variance and helps prevent overfitting.

  - **Boosting** builds models sequentially, where each new model learns from the mistakes of the previous ones. It reduces bias and focuses on hard-to-predict examples.





3.  What is the key idea behind AdaBoost
  - The key idea behind **AdaBoost (Adaptive Boosting)** is to combine multiple weak learners—typically simple decision trees—into a single strong learner by focusing more on the data points that previous models got wrong. It assigns higher weights to misclassified examples so that subsequent models pay more attention to them, improving accuracy over iterations.



4.  Explain the working of AdaBoost with an example
  - Here's a simple breakdown of how **AdaBoost** works using an example:

  1. Start with a dataset (e.g., classifying emails as spam or not spam).
  2. Assign equal weights to all data points.
  3. Train a weak learner (like a small decision tree).
  4. Identify which emails were misclassified.
  5. Increase the weights of those misclassified examples.
  6. Train the next weak learner, now focusing more on the hard examples.
  7. Repeat for several rounds, each time adjusting weights.
  8. Combine all the weak learners into a final strong classifier—more accurate and robust.

  So, AdaBoost keeps learning from mistakes and improves step by step.



5.  What is Gradient Boosting, and how is it different from AdaBoost
  - **Gradient Boosting** is a boosting technique where models are built sequentially, and each new model tries to correct the **errors (residuals)** of the previous one by optimizing a **loss function** using gradient descent. Like AdaBoost, it combines weak learners to form a strong one—but the method of improvement is different.

  **Key difference:**  
  - **AdaBoost** adjusts weights of misclassified examples.  
  - **Gradient Boosting** minimizes a loss function by fitting the next model to the residuals (errors) of the combined previous models.

  In short, **AdaBoost focuses on misclassified points**, while **Gradient Boosting focuses on reducing prediction errors overall**.




6.  What is the loss function in Gradient Boosting
  - In Gradient Boosting, the **loss function** measures how far off the model's predictions are from the actual values. The goal is to minimize this loss with each new model added.

  Common loss functions include:
  - **Mean Squared Error (MSE)** for regression tasks.
  - **Log Loss (Binary Cross-Entropy)** for binary classification.
  - **Multi-class Log Loss** for multi-class problems.

  Each new learner is trained to predict the **negative gradient** of the loss function (i.e., the direction of steepest descent), improving the overall model.




7.  How does XGBoost improve over traditional Gradient Boosting
  - **XGBoost** (Extreme Gradient Boosting) improves over traditional Gradient Boosting with several enhancements:

  - **Regularization:** Adds L1 and L2 regularization to control overfitting.
  - **Parallelization:** Uses parallel processing for faster training.
  - **Handling Missing Values:** Smartly handles missing data without needing imputation.
  - **Tree Pruning:** Uses a technique called “max depth” rather than growing trees greedily.
  - **Scalability:** Highly optimized for large-scale datasets.

  In short, XGBoost is faster, more accurate, and better at avoiding overfitting than standard Gradient Boosting.




8.  What is the difference between XGBoost and CatBoost
  - **XGBoost** and **CatBoost** are both powerful gradient boosting frameworks, but they differ in a few key ways:

  - **Categorical Features:**  
    - *XGBoost* requires manual encoding (like one-hot or label encoding).  
    - *CatBoost* automatically handles categorical data efficiently.

  - **Training Time:**  
    - *CatBoost* often trains faster on datasets with many categorical features.  
    - *XGBoost* is faster on purely numerical data due to optimized implementations.

  - **Accuracy:**  
    - *CatBoost* can outperform XGBoost on datasets with many categorical variables.  
    - *XGBoost* is highly accurate and widely adopted, especially in numerical-heavy tasks.

  In short: **CatBoost excels with categorical data**, while **XGBoost shines with numerical features and scalability**.




9.  What are some real-world applications of Boosting techniques
  - Boosting techniques are widely used in real-world tasks where high prediction accuracy is crucial. Here are some quick examples:

  - **Fraud Detection:** Banks use boosting to detect unusual transactions and prevent fraud.
  - **Medical Diagnosis:** Helps in predicting diseases from medical data with high precision.
  - **Spam Filtering:** Boosting models classify emails as spam or not with great accuracy.
  - **Customer Churn Prediction:** Businesses use it to identify customers likely to leave.
  - **Recommendation Systems:** Enhances personalization in platforms like e-commerce and streaming services.
  - **Credit Scoring:** Lenders assess creditworthiness using boosted models on financial data.

  It’s like giving your model superpowers—learning from mistakes to make sharper predictions.



10.  How does regularization help in XGBoost
  - **Regularization** in XGBoost helps **prevent overfitting** by adding penalties to complex models. It controls the model’s flexibility by discouraging overly deep or complicated trees that may fit the training data too well but fail on new data.

  There are two types used:
  - **L1 (Lasso)** regularization: Encourages sparsity by shrinking some weights to zero.
  - **L2 (Ridge)** regularization: Smooths the weights to keep them small but non-zero.

  Together, they make the model more **robust and generalizable**.




11.  What are some hyperparameters to tune in Gradient Boosting models
  - Here are key hyperparameters to tune in Gradient Boosting models:

  - **Learning Rate:** Controls how much each tree contributes; lower values need more trees.
  - **Number of Estimators:** Total trees in the model; more trees can boost accuracy but may increase overfitting.
  - **Max Depth:** Limits tree depth to control complexity and overfitting.
  - **Min Samples Split/Leaf:** Minimum data needed to split a node or be in a leaf; helps reduce overfitting.
  - **Subsample:** Fraction of data used per tree; adding randomness helps generalization.
  - **Loss Function:** Specifies the error measure (e.g., MSE, log loss).
  - **Regularization Parameters:** In XGBoost, includes `alpha` (L1) and `lambda` (L2) to penalize complexity.

  Tuning these well can significantly improve model performance.




12. What is the concept of Feature Importance in Boosting
  - **Feature Importance** in boosting tells us which input features (like age, salary, etc.) are most useful in making predictions. Boosting algorithms, like XGBoost or CatBoost, calculate how often and how effectively each feature is used to split data in decision trees.

  In short, **it highlights the features that contribute most to model accuracy**, helping with model interpretation and feature selection.



13.  Why is CatBoost efficient for categorical data?
  - CatBoost is efficient for categorical data because it **handles categorical features automatically**, without the need for manual preprocessing like one-hot or label encoding. It uses a technique called **ordered boosting**, which reduces overfitting and data leakage by using a special way of converting categorical values into numbers based on target statistics—while avoiding bias.

  This smart handling leads to better accuracy, faster training, and less manual work when working with categorical-heavy datasets.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# 1.  Train an AdaBoost Classifier on a sample dataset and print model accuracy


from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the AdaBoost Classifier
ada_classifier = AdaBoostClassifier(n_estimators=50, random_state=42)
ada_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = ada_classifier.predict(X_test)

# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
```

```python
# 2.  Train an AdaBoost Regressor and evaluate performance using Mean Absolute Error (MAE)


from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Load the California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the AdaBoost Regressor
ada_regressor = AdaBoostRegressor(n_estimators=50, random_state=42)
ada_regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = ada_regressor.predict(X_test)

# Calculate and print the Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")
```

```python
# 3.  Train a Gradient Boosting Classifier on the Breast Cancer dataset and print feature importance


from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_classifier.fit(X_train, y_train)

# Make predictions and calculate accuracy (for reference)
y_pred = gb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}\n")

# Get feature importance
feature_importance = gb_classifier.feature_importances_

# Create a DataFrame for better visualization
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

# Print feature importance
print("Feature Importance:")
print(importance_df)
```

```python
 # 4.  Train a Gradient Boosting Regressor and evaluate using R-Squared Score



from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Load the California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_regressor.fit(X_train, y_train)

# Make predictions and calculate R² score for Gradient Boosting
y_pred_gb = gb_regressor.predict(X_test)
r2_gb = r2_score(y_test, y_pred_gb)

# Initialize and train the AdaBoost Regressor for comparison
ada_regressor = AdaBoostRegressor(n_estimators=50, random_state=42)
ada_regressor.fit(X_train, y_train)

# Make predictions and calculate R² score for AdaBoost
y_pred_ada = ada_regressor.predict(X_test)
r2_ada = r2_score(y_test, y_pred_ada)

# Print R² scores
print(f"Gradient Boosting Regressor R² Score: {r2_gb:.4f}")
print(f"AdaBoost Regressor R² Score: {r2_ada:.4f}")
print(f"R² Score Difference (Gradient Boosting - AdaBoost): {r2_gb - r2_ada:.4f}")
```

```python
# 5.   Train an XGBoost Classifier on a dataset and compare accuracy with Gradient Boosting


from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_classifier.fit(X_train, y_train)

# Make predictions and calculate accuracy for Gradient Boosting
y_pred_gb = gb_classifier.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)

# Initialize and train the XGBoost Classifier
xgb_classifier = xgb.XGBClassifier(n_estimators=100, random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_classifier.fit(X_train, y_train)

# Make predictions and calculate accuracy for XGBoost
y_pred_xgb = xgb_classifier.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)

# Print accuracies
print(f"Gradient Boosting Classifier Accuracy: {accuracy_gb:.4f}")
print(f"XGBoost Classifier Accuracy: {accuracy_xgb:.4f}")
print(f"Accuracy Difference (XGBoost - Gradient Boosting): {accuracy_xgb - accuracy_gb:.4f}\n")

# Get feature importance for XGBoost
feature_importance = xgb_classifier.feature_importances_

# Create a DataFrame for feature importance
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

# Print feature importance for XGBoost
print("XGBoost Feature Importance:")
print(importance_df)
```

```python
# 6.  Train a CatBoost Classifier and evaluate using F1-Score


from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score


# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the CatBoost Classifier
cat_classifier = CatBoostClassifier(n_estimators=100, random_state=42, verbose=0)
cat_classifier.fit(X_train, y_train)

# Make predictions and calculate F1-Score and accuracy for CatBoost
y_pred_cat = cat_classifier.predict(X_test)
f1_cat = f1_score(y_test, y_pred_cat)
accuracy_cat = accuracy_score(y_test, y_pred_cat)

# Train Gradient Boosting Classifier for comparison
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_classifier.fit(X_train, y_train)
y_pred_gb = gb_classifier.predict(X_test)
f1_gb = f1_score(y_test, y_pred_gb)
accuracy_gb = accuracy_score(y_test, y_pred_gb)

# Train XGBoost Classifier for comparison
xgb_classifier = xgb.XGBClassifier(n_estimators=100, random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_classifier.fit(X_train, y_train)
y_pred_xgb = xgb_classifier.predict(X_test)
f1_xgb = f1_score(y_test, y_pred_xgb)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)

# Print F1-Scores and accuracies
print(f"CatBoost Classifier F1-Score: {f1_cat:.4f}")
print(f"CatBoost Classifier Accuracy: {accuracy_cat:.4f}\n")
print(f"Gradient Boosting Classifier F1-Score: {f1_gb:.4f}")
print(f"Gradient Boosting Classifier Accuracy: {accuracy_gb:.4f}\n")
print(f"XGBoost Classifier F1-Score: {f1_xgb:.4f}")
print(f"XGBoost Classifier Accuracy: {accuracy_xgb:.4f}\n")
print(f"F1-Score Difference (CatBoost - Gradient Boosting): {f1_cat - f1_gb:.4f}")
print(f"F1-Score Difference (CatBoost - XGBoost): {f1_cat - f1_xgb:.4f}")
```

```python
# 7.   Train an XGBoost Regressor and evaluate using Mean Squared Error (MSE)


from xgboost import XGBRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


# Load the California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the XGBoost Regressor
xgb_regressor = XGBRegressor(n_estimators=100, random_state=42)
xgb_regressor.fit(X_train, y_train)

# Make predictions and calculate MSE for XGBoost
y_pred_xgb = xgb_regressor.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)

# Initialize and train the AdaBoost Regressor for comparison
ada_regressor = AdaBoostRegressor(n_estimators=50, random_state=42)
ada_regressor.fit(X_train, y_train)

# Make predictions and calculate MSE for AdaBoost
y_pred_ada = ada_regressor.predict(X_test)
mse_ada = mean_squared_error(y_test, y_pred_ada)

# Print MSE results
print(f"XGBoost Regressor MSE: {mse_xgb:.4f}")
print(f"AdaBoost Regressor MSE: {mse_ada:.4f}")
print(f"MSE Difference (XGBoost - AdaBoost): {mse_xgb - mse_ada:.4f}")
```

```python
# 9.  Train a Gradient Boosting Regressor and plot learning curves



from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import mean_squared_error


# Load the California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Compute learning curves
train_sizes, train_scores, val_scores = learning_curve(
    gb_regressor, X_train, y_train, cv=5, scoring='neg_mean_squared_error',
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)

# Calculate mean and standard deviation for training and validation scores
train_scores_mean = -train_scores.mean(axis=1)  # Convert to positive MSE
train_scores_std = train_scores.std(axis=1)
val_scores_mean = -val_scores.mean(axis=1)  # Convert to positive MSE
val_scores_std = val_scores.std(axis=1)

# Plot learning curves
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label='Training MSE', color='blue', marker='o')
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color='blue')
plt.plot(train_sizes, val_scores_mean, label='Validation MSE', color='orange', marker='o')
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1, color='orange')

plt.title('Gradient Boosting Regressor Learning Curves')
plt.xlabel('Training Set Size')
plt.ylabel('Mean Squared Error')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()

# Train the model on the full training set and print test MSE
gb_regressor.fit(X_train, y_train)
y_pred = gb_regressor.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test Set MSE: {test_mse:.4f}")

# Display the plot
plt.show()
```

```python
# 12.  Train an AdaBoost Classifier with different numbers of estimators and compare accuracy

from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a range of n_estimators to test
n_estimators_list = [10, 50, 100, 200, 300]

# Store results
results = []

# Train AdaBoost Classifier for each number of estimators
for n_estimators in n_estimators_list:
    # Initialize and train the classifier
    ada_classifier = AdaBoostClassifier(n_estimators=n_estimators, random_state=42)
    ada_classifier.fit(X_train, y_train)

    # Make predictions and calculate accuracy
    y_pred = ada_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    # Store results
    results.append({'n_estimators': n_estimators, 'Accuracy': accuracy})

# Create a DataFrame for results
results_df = pd.DataFrame(results)

# Print results
print("AdaBoost Classifier Accuracy for Different Numbers of Estimators:")
print(results_df)

# Visualize accuracy vs. number of estimators
plt.figure(figsize=(8, 6))
sns.lineplot(x='n_estimators', y='Accuracy', data=results_df, marker='o')
plt.title('AdaBoost Classifier Accuracy vs. Number of Estimators')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# 13. Train a Gradient Boosting Classifier and visualize the ROC curve





from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, accuracy_score


# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
class_names = data.target_names  # ['malignant', 'benign']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_classifier.fit(X_train, y_train)

# Make predictions and calculate accuracy (for reference)
y_pred = gb_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")

# Get predicted probabilities for the positive class (benign)
y_pred_proba = gb_classifier.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)  # Diagonal line (random classifier)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Gradient Boosting Classifier ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()

# Print AUC score
print(f"AUC Score: {roc_auc:.4f}")
```

```python
# 14.  Train an XGBoost Regressor and tune the learning rate using GridSearchCV


from xgboost import XGBRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error


# Load the California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for learning rate
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3]
}

# Initialize the XGBoost Regressor
xgb_regressor = XGBRegressor(n_estimators=100, random_state=42)

# Perform GridSearchCV to tune learning rate
grid_search = GridSearchCV(estimator=xgb_regressor, param_grid=param_grid,
                           scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best model and its parameters
best_xgb = grid_search.best_estimator_
best_learning_rate = grid_search.best_params_['learning_rate']

# Make predictions with the best model and calculate MSE
y_pred_best = best_xgb.predict(X_test)
mse_best = mean_squared_error(y_test, y_pred_best)

# Train a default XGBoost Regressor for comparison (default learning_rate=0.3)
default_xgb = XGBRegressor(n_estimators=100, random_state=42)
default_xgb.fit(X_train, y_train)
y_pred_default = default_xgb.predict(X_test)
mse_default = mean_squared_error(y_test, y_pred_default)

# Create a DataFrame for GridSearchCV results
results_df = pd.DataFrame(grid_search.cv_results_)
results_df = results_df[['param_learning_rate', 'mean_test_score', 'std_test_score']]
results_df['mean_test_score'] = -results_df['mean_test_score']  # Convert to positive MSE
results_df = results_df.rename(columns={'param_learning_rate': 'Learning Rate',
                                       'mean_test_score': 'Mean CV MSE',
                                       'std_test_score': 'Std CV MSE'})

# Print results
print("GridSearchCV Results for Learning Rate Tuning:")
print(results_df)
print(f"\nBest Learning Rate: {best_learning_rate}")
print(f"Best Model Test MSE: {mse_best:.4f}")
print(f"Default Model Test MSE (learning_rate=0.3): {mse_default:.4f}")
print(f"MSE Improvement (Default - Best): {mse_default - mse_best:.4f}")
```

```python
# 15.  Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting



from catboost import CatBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.utils import resample

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
class_names = data.target_names  # ['malignant', 'benign']

# Create an imbalanced dataset by undersampling the majority class (benign, y=1)
df = pd.DataFrame(X)
df['target'] = y
majority = df[df['target'] == 1]  # Benign (357 samples)
minority = df[df['target'] == 0]  # Malignant (212 samples)

# Adjust the number of majority samples to achieve ~80:20 ratio, respecting available data
n_majority = min(len(majority), int(len(minority) * 4))  # Cap at available majority samples
majority_downsampled = resample(majority, replace=False, n_samples=n_majority, random_state=42)
imbalanced_df = pd.concat([majority_downsampled, minority])
X_imbalanced = imbalanced_df.drop('target', axis=1).values
y_imbalanced = imbalanced_df['target'].values

# Print class distribution for verification
print(f"Imbalanced dataset class distribution: Malignant={np.sum(y_imbalanced == 0)}, Benign={np.sum(y_imbalanced == 1)}")
print(f"Ratio (Benign:Malignant): {np.sum(y_imbalanced == 1) / np.sum(y_imbalanced == 0):.2f}:1\n")

# Split the imbalanced dataset
X_train, X_test, y_train, y_test = train_test_split(X_imbalanced, y_imbalanced, test_size=0.3, random_state=42)

# Calculate class weights for the imbalanced dataset
class_counts = np.bincount(y_train)
class_weights = [class_counts.sum() / (2.0 * class_counts[0]), class_counts.sum() / (2.0 * class_counts[1])]

# Initialize and train standard CatBoost Classifier
cat_standard = CatBoostClassifier(n_estimators=100, random_state=42, verbose=0)
cat_standard.fit(X_train, y_train)

# Initialize and train weighted CatBoost Classifier
cat_weighted = CatBoostClassifier(n_estimators=100, random_state=42, verbose=0, class_weights=class_weights)
cat_weighted.fit(X_train, y_train)

# Make predictions and evaluate both models
y_pred_standard = cat_standard.predict(X_test)
y_pred_weighted = cat_weighted.predict(X_test)

accuracy_standard = accuracy_score(y_test, y_pred_standard)
f1_standard = f1_score(y_test, y_pred_standard)
accuracy_weighted = accuracy_score(y_test, y_pred_weighted)
f1_weighted = f1_score(y_test, y_pred_weighted)

# Compute confusion matrices
cm_standard = confusion_matrix(y_test, y_pred_standard)
cm_weighted = confusion_matrix(y_test, y_pred_weighted)

# Print results
print("Standard CatBoost Classifier (No Class Weighting):")
print(f"Accuracy: {accuracy_standard:.4f}")
print(f"F1-Score: {f1_standard:.4f}\n")
print("Weighted CatBoost Classifier (With Class Weighting):")
print(f"Accuracy: {accuracy_weighted:.4f}")
print(f"F1-Score: {f1_weighted:.4f}\n")
print(f"F1-Score Difference (Weighted - Standard): {f1_weighted - f1_standard:.4f}")

# Visualize confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Standard model confusion matrix
sns.heatmap(cm_standard, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names, ax=axes[0])
axes[0].set_title('Standard CatBoost Confusion Matrix')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')

# Weighted model confusion matrix
sns.heatmap(cm_weighted, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names, ax=axes[1])
axes[1].set_title('Weighted CatBoost Confusion Matrix')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')

plt.tight_layout()
plt.show()
```

```python
# 16.  Train an AdaBoost Classifier and analyze the effect of different learning rates



from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a range of learning rates to test
learning_rates = [0.01, 0.1, 0.5, 1.0, 2.0]

# Store results
results = []

# Train AdaBoost Classifier for each learning rate
for lr in learning_rates:
    # Initialize and train the classifier
    ada_classifier = AdaBoostClassifier(n_estimators=50, learning_rate=lr, random_state=42)
    ada_classifier.fit(X_train, y_train)

    # Make predictions
    y_pred = ada_classifier.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Store results
    results.append({'Learning Rate': lr, 'Accuracy': accuracy, 'F1-Score': f1})

# Create a DataFrame for results
results_df = pd.DataFrame(results)

# Print results
print("AdaBoost Classifier Performance for Different Learning Rates:")
print(results_df)

# Visualize accuracy and F1-Score vs. learning rate
plt.figure(figsize=(10, 6))
plt.plot(results_df['Learning Rate'], results_df['Accuracy'], marker='o', label='Accuracy', color='blue')
plt.plot(results_df['Learning Rate'], results_df['F1-Score'], marker='s', label='F1-Score', color='orange')
plt.title('AdaBoost Classifier: Accuracy and F1-Score vs. Learning Rate')
plt.xlabel('Learning Rate')
plt.ylabel('Score')
plt.xscale('linear')  # Linear scale for better interpretability
plt.grid(True)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# 17.  Train an XGBoost Classifier for multi-class classification and evaluate using log-loss.


from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, accuracy_score

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
class_names = data.target_names  # ['setosa', 'versicolor', 'virginica']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the XGBoost Classifier for multi-class classification
xgb_classifier = XGBClassifier(n_estimators=100, random_state=42, objective='multi:softprob', eval_metric='mlogloss')
xgb_classifier.fit(X_train, y_train)

# Make predictions
y_pred = xgb_classifier.predict(X_test)  # Predicted class labels
y_pred_proba = xgb_classifier.predict_proba(X_test)  # Predicted probabilities

# Calculate log-loss and accuracy
logloss = log_loss(y_test, y_pred_proba)
accuracy = accuracy_score(y_test, y_pred)

# Print results
print(f"XGBoost Classifier (Multi-Class) Performance:")
print(f"Log-Loss: {logloss:.4f}")
print(f"Accuracy: {accuracy:.4f}")

# Visualize log-loss and accuracy
metrics = pd.DataFrame({
    'Metric': ['Log-Loss', 'Accuracy'],
    'Value': [logloss, accuracy]
})

plt.figure(figsize=(8, 6))
sns.barplot(x='Metric', y='Value', data=metrics, palette='Blues')
plt.title('XGBoost Classifier: Log-Loss and Accuracy')
plt.ylabel('Score')
plt.ylim(0, max(logloss, accuracy) * 1.2)  # Adjust y-axis for visibility
for i, v in enumerate(metrics['Value']):
    plt.text(i, v + 0.02, f'{v:.4f}', ha='center')
plt.tight_layout()

# Display the plot
plt.show()
```
