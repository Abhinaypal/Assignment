### 8. What is a hash table, and how does it relate to dictionaries in Python?
  - Hash table in Python is a data sturcture which stores data by using a pair of values and keys. Each data value is allocated an exclusive key that is created by using the hash fucntions.
  - Dictionaries are implemented using hash tables. When a key-value pair is added to a dictionary, Pythong calculates the hash value of the key and uses it to determine the location to store the value. When retrieving a value, Python again calculates the hash value of the key and uses it to quickly locate the corresponding value. This implementation makes dictionaries very efficient for lookups, insertions, and deletions.

### 14. How does the "in" keyword work differently for lists and dictionaries?
  - When using the "in" keyword in Python, it checks if an element is present in a list based on its position (index), while in a dictionary, it checks if a specific key exists within the key-value pairs. In a list you are checking if a value is "in" the sequence, whereas in a dictionary, you are checking if a "key" is "in" the collection.


  my_list = ["apple", "banana", "orange"]

  print("apple" in my_list)  # Output: True (because "apple" is in the list)


  my_dict = {"fruit": "apple", "color": "red"}

  print("apple" in my_dict)  # Output: True (because "apple" is a key in the dictionary)

### 17.  Describe the time complexity of accessing elements in a dictionary.
  - The time complexity of accessing an element in a dictionary is O(1) on average, which means it takes constant time regardless of the dictionary's size. However, the worst-case time complexity can be O(n) if there are many collisions.



### 18.  In what situations are lists preferred over dictionaries?
  - Lists are preferred over dictionaries when the order of elements is crucial, you need to perform operations on sequential data, or when you want to easily add or remove elements from a collection in a specific order, as lists maintain the order of their elements while dictionaries prioritize fast lookups using keys.



### 19.  Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Dictionaries are considered unordered because they don't maintain the order in which items are added. This is because dictionaries are optimized for fast lookups based on keys.
