## Theory Questions


1. What is numpy, and why is it widely used in Python?
  - NumPy is a Python library used for numerical computing. It provides support for large multi-dimensional arrays and matrices, along with a collection of mathematical functions to perform operations on these arrays. NumPy is widely used because it offers efficient operations on arrays and matrices, which are essential for scientific and mathematical computing, data analysis, and machine learning. Its ability to interface with other libraries like Pandas and Scikit-learn enhances its utility even further.


2. How does broadcasting work in NumPy?
  - Broadcasting in NumPy allows operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger array. This feature is useful for performing element-wise operations without explicitly reshaping arrays, improving code efficiency and readability.


3.  What is a Pandas DataFrame?
  - A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure in Python. It is similar to a table or a spreadsheet, with rows and columns. Each column in a DataFrame can hold different types of data, such as integers, floats, or strings. DataFrames are widely used for data manipulation, analysis, and visualization, offering powerful tools for filtering, aggregating, and transforming data. They also support operations like merging, joining, and reshaping datasets, making them essential for data science and machine learning tasks.

6.  What are the differences between NumPy arrays and Python lists?
  - NumPy arrays are more efficient for numerical computations and can handle large data sets, offering element-wise operations and broadcasting. Python lists are more flexible in terms of data types but are slower for numerical tasks and don't support vectorized operations. NumPy arrays have fixed size, whereas Python lists can grow dynamically. Overall, NumPy arrays are preferred for scientific and mathematical applications, while Python lists are used for general-purpose programming.


7.  What is a heatmap, and when should it be used?
  - A heatmap is a data visualization technique that uses colors to represent the values in a matrix or table. It is used to reveal patterns, correlations, and variations within the data, making it easier to spot trends and anomalies. Heatmaps are particularly useful in fields like biology, finance, and machine learning for visualizing complex datasets.


8.  What does the term “vectorized operation” mean in NumPy?
  - In NumPy, a vectorized operation refers to performing element-wise operations on entire arrays without using explicit loops. This approach leverages low-level optimizations and parallel processing, resulting in faster and more efficient computations compared to traditional looping methods.


9.  How does Matplotlib differ from Plotly?
  - Matplotlib and Plotly are both popular libraries for data visualization in Python, but they differ in several ways. Matplotlib is a static plotting library that offers extensive customization and is ideal for creating publication-quality plots. It is particularly well-suited for creating basic graphs and charts. Plotly, on the other hand, is an interactive plotting library that allows for dynamic and web-based visualizations. It is great for creating complex and interactive plots, dashboards, and data exploration tools. While Matplotlib excels in static plots, Plotly is preferred for interactive and visually engaging visualizations.


10.  What is the significance of hierarchical indexing in Pandas?
  - Hierarchical indexing (multi-level indexing) in Pandas organizes data using multiple keys, making it easier to analyze and manipulate complex datasets. It allows operations across different levels and simplifies grouping.

  - Example:
import pandas as pd

data = {'State': ['CA', 'CA', 'NY', 'NY'], 'City': ['San Francisco', 'Los Angeles', 'New York', 'Buffalo'], 'Population': [870000, 4000000, 8500000, 260000]}
df = pd.DataFrame(data)
df.set_index(['State', 'City'], inplace=True)

print(df)


This creates a DataFrame with a hierarchical index based on 'State' and 'City':

               Population
State City               
CA    San Francisco   870000
      Los Angeles   4000000
NY    New York     8500000
      Buffalo        260000


This example shows how hierarchical indexing allows easy access and analysis of data at both the state and city levels.

17.  Explain the difference between apply() and map() in Pandas?
  - The apply() function in Pandas is used to apply a function along an axis of a DataFrame (either rows or columns), making it versatile for various operations. 21.  Why is NumPy’s array slicing faster than Python’s list slicing?

This reduces overhead and enables faster computation compared to Python lists, which are more flexible but less optimized for numerical tasks.

```python
# 2.  Write a Pandas script to find the mean of a specific column in a DataFrame.


import pandas as pd

data = {'Column1': [10, 20, 30, 40, 50], 'Column2': [5, 15, 25, 35, 45]}
df = pd.DataFrame(data)

mean_value = df['Column1'].mean()
print(mean_value)
```

```python
# 4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'A': [1, 2, 3, 4, 5], 'B': [5, 4, 3, 2, 1], 'C': [2, 3, 4, 5, 6]}
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix with a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 5.  Generate a bar plot using Plotly.

import plotly.graph_objects as go

# Sample data
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 20, 15, 25, 30]

# Create a bar plot
fig = go.Figure(data=[go.Bar(x=x, y=y)])

# Add labels and title
fig.update_layout(title='Bar Plot Example', xaxis_title='X-axis Label', yaxis_title='Y-axis Label')

# Show the plot
fig.show()
```

```python
# 7.  Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])

# Perform element-wise multiplication
result = array1 * array2
print(result)
```

```python
# 8.  Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [15, 25, 35, 45, 55]

# Create a line plot with multiple lines
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')

# Add labels, title, and legend
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot with Multiple Lines')
plt.legend()

# Show the plot
plt.show()
```

```python
# 9.  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.


import pandas as pd

# Create a DataFrame
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Filter rows where column 'A' values are greater than a threshold (e.g., 3)
threshold = 3
filtered_df = df[df['A'] > threshold]
print(filtered_df)
```

```python
# 13.  Create a 3D scatter plot using Plotly.


import plotly.graph_objects as go

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
z = [5, 15, 25, 35, 45]

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5))])

# Add labels and title
fig.update_layout(title='3D Scatter Plot Example', scene=dict(
    xaxis_title='X-axis Label',
    yaxis_title='Y-axis Label',
    zaxis_title='Z-axis Label'
))

# Show the plot
fig.show()
```
